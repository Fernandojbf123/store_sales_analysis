```python
# Importar librerías y leer los archivos de datos
import pandas as pd
df1 = pd.read_csv("Datos_Ventas_Tienda.csv")
df2 = pd.read_csv("Datos_Ventas_Tienda2.csv")
df1.head(10)
```

```python
# Mostrar el segundo DataFrame leído
df2.head(10)
```

```python
# Unir ambos DataFrames en uno solo para el análisis
df = pd.concat([df1,df2])
```

```python
# Convertir la columna 'Fecha' al formato datetime para facilitar el análisis por fechas
df["Fecha"] = pd.to_datetime(df["Fecha"],format="%m/%d/%Y")
```

```python
# ¿Cuál es el producto más vendido?
# Calcular el producto más vendido sumando la cantidad por producto
productos_mas_vendidos = df.groupby(["Producto"])["Cantidad"].sum()
producto_mas_vendido = productos_mas_vendidos.idxmax()
cantidad = productos_mas_vendidos.max()
print(f"El producto más vendido fue {producto_mas_vendido} y se vendieron {cantidad} unidades")
```

```python
# ¿Cuál es el mes con más ventas?
# Extraer el año y el mes de la fecha para análisis temporal
df["Año"] = df["Fecha"].dt.strftime("%Y")
df["Mes"] = df["Fecha"].dt.strftime("%m")
df.head(10)
```

```python
# Calcular el mes con más ventas sumando el total por año y mes
sum_ventas_mes = df.groupby(["Año","Mes"])["Total Venta"].sum()
sum_ventas_mes_ordenado = sum_ventas_mes.sort_values(ascending=False)
anio_mes_de_mas_ventas = sum_ventas_mes_ordenado.keys()[0]
cantidad_vendida = sum_ventas_mes_ordenado.iloc[0]
print(f"El mes con mas ventas fue el {anio_mes_de_mas_ventas[1]} con un total de {cantidad_vendida} pesos")
```

```python
# Guardar el DataFrame final con los resultados en un archivo CSV
df.to_csv('output_analysis.csv',index=False)
```
